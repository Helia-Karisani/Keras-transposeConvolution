```python
import scipy
import matplotlib
import tensorflow as tf

print("scipy:", scipy.__version__)
print("matplotlib:", matplotlib.__version__)
print("tensorflow:", tf.__version__)
```

I use numpy and matplotlib for data manipulation and visualization.

```python
import warnings
warnings.simplefilter('ignore')

from tensorflow.keras.models import Model 
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, UpSampling2D 
import numpy as np 
import matplotlib.pyplot as plt 
```

define input layer

```python
input_layer = Input(shape=(28, 28, 1))
```

adding transpose convolutional and convolutional layers

```python
conv_layer = Conv2D(filters=32, kernel_size=(3, 3), activation='relu', padding='same')(input_layer) 

transpose_conv_layer = Conv2DTranspose(filters=1, kernel_size=(3, 3), activation='sigmoid', padding='same')(conv_layer) 
```

creating the model

```python
model = Model(inputs=input_layer, outputs=transpose_conv_layer)
```

compile the model

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

train the model<br>

X_train will be the placeholder for the actual training data<br>
y_train is the same as X_train for image reconstruction.

```python
# generate synthetic training data 

X_train = np.random.rand(1000, 28, 28, 1) 

y_train = X_train # For reconstruction, the target is the input 

history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2) 
```

evaluate the model

```python
 # generate synthetic test data 
# Create 200 random fake images of size 28x28x1
X_test = np.random.rand(200, 28, 28, 1) 

y_test = X_test 

loss = model.evaluate(X_test, y_test) 

print(f'Test loss: {loss}')
```

visualize some reconstructed images

```python
# Predict on test data 
y_pred = model.predict(X_test) 

# Plot some sample images 

n = 10 # Number of samples to display 

plt.figure(figsize=(20, 4))

for i in range(n): 

    # Display original 
    ax = plt.subplot(2, n, i + 1) 
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.title("Original") 
    plt.axis('off') 
    # Display reconstruction 
    ax = plt.subplot(2, n, i + 1 + n) 
    plt.imshow(y_pred[i].reshape(28, 28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')

plt.show() 
```
